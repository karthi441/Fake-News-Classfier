```python
def removeShortWords(text):
  res = ""
  for i in text.split(' '):
    if len(i) <= 2:
      continue
    else:
      res += i
    res += ' '
  return res
```

# Logistic Regression

```python
!pip install yellowbrick
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,accuracy_score
from sklearn.model_selection import KFold
import pickle
from yellowbrick.model_selection import learning_curve

kf = KFold(n_splits=5, random_state=0, shuffle=True)

# K-Fold on the training set for train-val split
best_clf_log = None
best_clf_log_score = -2147483648  #INT_MIN
for train_index, test_index in kf.split(X_train):
  x_train, x_val = X_train[train_index], X_train[test_index]
  y_train, y_val = Y_train[train_index], Y_train[test_index]

  clf_log = LogisticRegression(max_iter=1000)
  clf_log.fit(x_train,y_train)

  y_pred = clf_log.predict(x_val)
  cr = classification_report(y_val,y_pred)
  acc = accuracy_score(y_val, y_pred)
  if(best_clf_log_score < acc):
    best_clf_log_score = acc
    best_clf_log = clf_log

  print(cr);
  print()

# Saving the best model using pickle
pickle.dump(best_clf_log, open('clf_log.sav', 'wb'))
```

```python
# Final accuracy on the real testing set i.e X_test, Y_test
y_test_pred = best_clf_log.predict(X_test)
cr = classification_report(Y_test,y_test_pred)
print(cr);
```

```python
from sklearn.metrics import plot_roc_curve
plot_roc_curve(best_clf_log, X_test, Y_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(Y_test, y_test_pred, labels=best_clf_log.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_clf_log.classes_)
disp.plot()
plt.show()
```
